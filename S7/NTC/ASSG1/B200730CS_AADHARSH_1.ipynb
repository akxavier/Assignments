```python
def affine_encrypt(ptext, k1, k2):
    ctext = ""
    for ch in ptext:
        if ch == " ":
            ctext += ch
        else:
            ctext += chr(((ord(ch) - ord('a')) * k1 + k2) % 26 + ord('A'))
    return ctext
```

```python
affine_encrypt('hello', 7, 2)
```

```python
# Decrypt Affine Cipher
def affine_decrypt(ctext, k1, k2):
    ptext = ""
    for i in range(1, 26, 2):
        if (k1*i % 26) == 1:
            inverse = i
            break
    for ch in ctext:
        if ch == " ":
            ptext += " "
        else:
            ptext += chr(((ord(ch) - ord('A') - k2) * inverse) % 26 + ord('a'))
    return ptext
```

```python
affine_decrypt('ZEBBW', 7, 2)
```

```python
# Cryptanalysis
ctext = 'ZEBBW'
for i in range(1, 26, 2):
    if i != 13:
        for j in range(26):
            print("k1:", i, "k2:", j, "Plaintext =", affine_decrypt(ctext, i, j))
```
